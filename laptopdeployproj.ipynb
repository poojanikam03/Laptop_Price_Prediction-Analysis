```python
#Will Filter AMD c.p.u w.r.t to the series
#AMD A9-Series 9410	>> AMD A-Series
#AMD E-Series 9000>>AMD E-Series
#AMD FX 8800P>>AMD F-Series
#AMD Ryzen 1600	>>AMD Ryzen-Series
```

```python
def fetch_pro(text):
  if text=='Intel Core i7' or text=='Intel Core i3' or text=='Intel Core i5':
    return text
  elif text.split()[0]=='Intel':
    return "Other Intel Processor"
  else:
    if text[4]=='E':
      return "AMD E-Series"
    elif text[4]=='A':
      return "AMD A-Series"
    elif text[4]=='F':
      return "AMD F-Series"
    else:
      return "AMD Ryzen Series"



```

```python
df['Cpu']=df['Cpu'].apply(fetch_pro) #Applying the function on cpu column
```

```python
#wherever we have Windows 10 or Windows 10 S >> Windows 10
#Wherever we have macOS or Mac OS X	>>mac OS
#Linux>>Linux
#Windows 7>> Windows 7
#No OS>>No OS
#Chrome OS	>> Chrome OS
```

```python
def os_text(text):
  if text=='Windows 10' or text=='Windows 10 S':
    return 'Windows 10'
  elif text=='macOS' or text=='Mac OS X':
    return 'mac OS'
  elif text=='Linux':
    return 'Linux'
  elif text=='Windows 7':
    return 'Windows 7'
  elif text=='Chrome OS':
    return 	'Chrome OS'
  else:
    return 'No OS'



```

```python
os_text('Mac OS X')
```

```python
df['OpSys']=df['OpSys'].apply(os_text)
```
